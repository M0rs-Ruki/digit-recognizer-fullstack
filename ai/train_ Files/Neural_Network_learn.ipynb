```python
# Three Weights Nural Dynamic 

inputs = [1, 2, 3, 2.5]

weight = [[0.2, 0.8, -0.5, 1.0],
          [0.5, -0.91, 0.26, -0.5],
          [-0.26, -0.27, 0.17, 0.87]]

bias = [2, 3, 0.5]

layer_output = []
for neuron_weights, neuron_bias in zip(weight, bias):
    neuron_output = 0
    for n_input, w in zip(inputs, neuron_weights):
        neuron_output += n_input * w
    neuron_output += neuron_bias
    layer_output.append(neuron_output)

layer_output
```

```python
import numpy as np

# Set random seed for reproducibility
np.random.seed(0)

# Input data
X = [[1, 2, 3, 2.5],
     [2.0, 5.0, -1.0, 2.0],
     [-1.5, 2.7, 3.3, -0.8]]

# Dense layer class
class LayerDense:
    def __init__(self, n_inp, n_neu):
        # Initialize weights and biases
        self.weights = 0.10 * np.random.randn(n_inp, n_neu)
        self.biases = np.zeros((1, n_neu))

    def forward(self, inputs):
        # Perform forward pass
        self.output = np.dot(inputs, self.weights) + self.biases

# Define layers
layer_1 = LayerDense(4, 5)  # 4 inputs, 5 neurons
layer_2 = LayerDense(5, 2)  # 5 inputs (output of layer_1), 2 neurons

# Forward pass for layer_1
layer_1.forward(X)

# Forward pass for layer_2 using layer_1's output
layer_2.forward(layer_1.output)

# Print the output of layer_2
("Layer 1 Output:")
(layer_1.output)

("\nLayer 2 Output:")
(layer_2.output)
```

```python
import numpy as np
import pandas as pd

np.random.seed(0)

# Dense layer class
class LayerDense:
    def __init__(self, n_inp, n_neu):
        # Initialize weights and biases
        self.weights = 0.10 * np.random.randn(n_inp, n_neu)
        self.biases = np.zeros((1, n_neu))
    def forward(self, inputs):
        # Perform forward pass
        self.output = np.dot(inputs, self.weights) + self.biases


class Activation_Relu:
    def forward(self, inputs):
        self.inputs = inputs
        self.output = np.maximum(0, inputs)


class Activation_Softmax:
    def forward(self, inputs):
        exp_values = np.exp(inputs - np.max(inputs, axis=1, keepdims=True))
        probabilities = exp_values / np.sum(exp_values, axis=1, keepdims=True)
        self.output = probabilities
        

class Loss:
    def calculate(self, output, y):
        sample_losses = self.forward(output, y)
        data_loss = np.mean(sample_losses)
        return data_loss

class Loss_CategoricalCrossentropy(Loss):
    def forward(self, y_pred, y_true):
        samples = len(y_pred)
        y_pred_clipped = np.clip(y_pred, 1e-7, 1-1e-7)
        
        if len(y_true.shape) == 1:
            correct_confidences = y_pred_clipped[range(samples), y_true]

        elif len(y_true.shape) == 2:
            correct_confidences = np.sum(y_pred_clipped * y_true, axis=1)

        negative_log_likelihoods = -np.log(correct_confidences)
        return negative_log_likelihoods


# Load the data
data = pd.read_csv('C:/Users/anupp/OneDrive/Documents/Python/Projectsss/digit-recognizer/train.csv')

# Separate features (X) and labels (y)
X = data.iloc[:, 1:].values  # All rows, columns from index 1 onward
y = data.iloc[:, 0].values   # All rows, first column (labels)

# Initialize layers
densel = LayerDense(X.shape[1], 128)  # Number of features in X, 128 neurons
activation1 = Activation_Relu()

dense2 = LayerDense(128, 10)  # Output layer, 10 neurons for 10 classes
activation2 = Activation_Softmax()

# Forward pass
densel.forward(X)
activation1.forward(densel.output)

dense2.forward(activation1.output)
activation2.forward(dense2.output)

# Print the first 5 outputs
print("First 5 outputs:\n", activation2.output[:5])

# Compute loss
loss_function = Loss_CategoricalCrossentropy()
loss = loss_function.calculate(activation2.output, y)

print("Loss:", loss)
```
